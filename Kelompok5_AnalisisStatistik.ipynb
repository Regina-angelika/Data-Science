# PENCILAN

Menggunakan Metode Turkey's Fences

```python
q1=fc[['overall_rating']].quantile(q=0.25)
q3=fc[['overall_rating']].quantile(q=0.75)
isol=(fc[['overall_rating']]<q1-1.5*(q3-q1))|(fc[['overall_rating']]>q1+1.5*(q3-q1))
exol=(fc[['overall_rating']]<q1-3*(q3-q1))|(fc[['overall_rating']]>q1+3*(q3-q1))
df2=fc[['name','overall_rating']].assign(isoutlier=isol,extremeOutlier=exol)
df2.head(60)
```

Berdasarkan hasil dari turkey's fences, dengan menggunakan data "overall_rating" sebagai atribut utama. Didapatkan hasil record sebanyak 60. Semua record untuk isooutlier bernilai "true". Sedangkan record untuk extremeOutlier hanya 1 record (pertama) yang bernilai "true", sisanya bernilai "false"
